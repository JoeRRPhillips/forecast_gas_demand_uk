```python
import numpy as np
import pandas as pd
import tensorflow as tf
from typing import Dict


# config: Dict = {
#     "train_data_path": "C:/Users/JPhillips/ldz/data/processed/train.csv",
#     "test_data_path": "C:/Users/JPhillips/ldz/data/processed/test.csv",
# }
config: Dict = {
    "epochs": 2,
    "batch_size": 2,
    "learning_rate": 3e-4,
    "train_data_path": "/mnt/c/Users/JPhillips/ldz/data/processed/train.csv",
    "test_data_path": "/mnt/c/Users/JPhillips/ldz/data/processed/test.csv",
}
```

```python
train_data = pd.read_csv(config["train_data_path"])
# train_data.head()
```

```python
features = train_data.copy()
labels = features.pop("demand")

# features.head(), labels.head()
```

```python
features = np.array(features)
labels = np.array(labels)
```

```python
train_dataset = tf.data.Dataset.from_tensor_slices((features, labels))
train_dataset = train_dataset.batch(config["batch_size"])
train_dataset
```

```python
from networks.base import LinearModel
```

```python
model = LinearModel(num_outputs=1)
model.compile(
    loss = "mse",
    optimizer = tf.keras.optimizers.Adam(learning_rate=config["learning_rate"])
)
```

```python
history = model.fit(
    train_dataset,
    epochs=config["epochs"],
)
```
